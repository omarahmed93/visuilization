```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as  plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

# Displot

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.distplot(tips['total_bill'],bins=10)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # kind,scatter,kde,hex
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='day',palette='dark:salmon_r')
```
